# Write a Program to implement data link layer farming method checksum

```python
# Function to perform binary addition with carry (wrap-around)
def binary_addition(a, b):
    max_len = max(len(a), len(b))
    a = a.zfill(max_len)
    b = b.zfill(max_len)

    result = ''
    carry = 0

    for i in range(max_len - 1, -1, -1):
        total = carry + int(a[i]) + int(b[i])
        result = str(total % 2) + result
        carry = total // 2

    # If carry is left, wrap around and add
    if carry:
        result = binary_addition(result, '1'.zfill(max_len))
    return result

# Function to calculate checksum (1's complement of sum)
def calculate_checksum(data_blocks):
    total = data_blocks[0]
    for block in data_blocks[1:]:
        total = binary_addition(total, block)

    # Invert bits to get checksum
    checksum = ''.join('1' if bit == '0' else '0' for bit in total)
    return checksum

# Function to verify checksum at receiver
def verify_checksum(data_blocks, received_checksum):
    total = data_blocks[0]
    for block in data_blocks[1:]:
        total = binary_addition(total, block)
    total = binary_addition(total, received_checksum)
    return all(bit == '1' for bit in total)

# Examples
examples = [
    ["10101010", "10101010", "11110000"],
    ["11001100", "11110000", "00001111"],
    ["00000000", "11111110", "10101010"],
    ["10011001", "01100110", "11000011"],
    ["11111111", "11111111", "11111111"]
]

# Run each example
for i, sender_data in enumerate(examples, start=1):
    print(f"\nExample {i}")
    print("Sender data blocks:")
    for block in sender_data:
        print(" ", block)

    checksum = calculate_checksum(sender_data)
    print("Calculated Checksum:", checksum)

    # Transmit data
    transmitted_data = sender_data.copy()
    transmitted_data.append(checksum)

    # Optionally inject error for Example 3
    if i == 3:
        print("Injecting error in Example 3...")
        # Flip 1 bit in the first block (simulate error)
        transmitted_data[0] = '10000000'  # changed from 00000000 to 10000000

    # Receiver checks
    print("Receiver side checking...")
    if verify_checksum(transmitted_data[:-1], transmitted_data[-1]):
        print("Data received correctly. No error detected.")
    else:
        print("Error in data transmission detected.")
```
