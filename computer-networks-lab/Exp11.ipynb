```python
import math

INF = math.inf

# Define nodes and weighted edges
nodes = ['A', 'B', 'C', 'D', 'E']
graph = {
    'A': {'B': 1, 'D': 4},
    'B': {'A': 1, 'C': 3, 'E': 1},
    'C': {'B': 3, 'E': 2},
    'D': {'A': 4, 'E': 2},
    'E': {'B': 1, 'C': 2, 'D': 2}
}

# Initialize distance tables and next-hop tables
dist = {u: {v: (0 if u == v else (graph[u][v] if v in graph[u] else INF)) for v in nodes} for u in nodes}
next_hop = {u: {v: (None if u == v else (v if v in graph[u] else None)) for v in nodes} for u in nodes}

print("Initial Distance Tables:")
for u in nodes:
    print(f"{u}: {dist[u]}")
print("\nStarting Distance Vector Updates...\n")

# Bellman-Ford style update until convergence
iteration = 0
while True:
    iteration += 1
    updated = False
    for u in nodes:
        for neighbor, link_cost in graph[u].items():
            for dest in nodes:
                new_cost = link_cost + dist[neighbor][dest]
                if new_cost < dist[u][dest]:
                    dist[u][dest] = new_cost
                    next_hop[u][dest] = neighbor
                    updated = True
    if not updated:
        break

print(f"Converged after {iteration} iterations.\n")

# Display routing tables
def print_routing_tables(dist, next_hop):
    for u in nodes:
        print(f"Routing Table for Node {u}")
        print(f"{'Destination':<12}{'Cost':<10}{'Next Hop':<10}")
        print("-" * 32)
        for v in nodes:
            cost = "INF" if dist[u][v] == INF else int(dist[u][v])
            hop = next_hop[u][v] if next_hop[u][v] is not None else "-"
            print(f"{v:<12}{cost:<10}{hop:<10}")
        print()

print_routing_tables(dist, next_hop)

# Function to reconstruct path
def get_path(src, dst):
    if dist[src][dst] == INF:
        return None
    path = [src]
    while src != dst:
        src = next_hop[src][dst]
        if src is None:
            return None
        path.append(src)
    return path

# Example paths
examples = [('A', 'C'), ('C', 'D'), ('D', 'B'), ('A', 'E')]
for s, d in examples:
    path = get_path(s, d)
    if path:
        print(f"Shortest Path {s} -> {d}: {' -> '.join(path)}  (Cost: {int(dist[s][d])})")
    else:
        print(f"No path found from {s} to {d}")
```
