```python
# To implement a Convolution Neural Network (CNN) for dog/cat classification problems using Keras.
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import os
import zipfile
import shutil
# from google.colab import drive
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.1,  # Reduced from 0.2
    zoom_range=0.1,   # Reduced from 0.2
    horizontal_flip=False # Changed to False
)

train_generator = train_datagen.flow_from_directory(
    # '/content/drive/MyDrive/MyDatasets/cats_and_dogs_filtered/cats_and_dogs_filtered/train', # REPLACE THIS PATH WITH THE PATH IN YOUR SYSTEM TO THE TRAINING DATASET
    # 'D:/AllSaradhiTeaching/DL_Lab2024/datasets/cats_and_dogs_filtered/train',
    'D:/DL Lab 2026/lab05/cats_and_dogs_filtered/train',
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    # '/content/drive/MyDrive/MyDatasets/cats_and_dogs_filtered/cats_and_dogs_filtered/validation', # REPLACE THIS PATH WITH THE PATH IN YOUR SYSTEM TO THE VALIDATION DATASET
    # 'D:/AllSaradhiTeaching/DL_Lab2024/datasets/cats_and_dogs_filtered/validation',
    'D:/DL Lab 2026/lab05/cats_and_dogs_filtered/validation',
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)
```

```python
model_seq = Sequential([
    Conv2D(16, (3, 3), activation='relu', input_shape=(150, 150, 3)), # Reduced filters from 32 to 16
    MaxPooling2D((2, 2)),

    Conv2D(32, (3, 3), activation='relu'), # Reduced filters from 64 to 32
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'), # Reduced filters from 128 to 64
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'), # Reduced filters from 128 to 64
    MaxPooling2D((2, 2)),

    Flatten(),
    Dense(256, activation='relu'), # Reduced units from 512 to 256
    Dense(1, activation='sigmoid')
])

model_seq.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_seq.summary()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense

# Define the input layer
inputs = Input(shape=(150, 150, 3))

# First convolutional block
x = Conv2D(16, (3, 3), activation='relu')(inputs) # Reduced filters from 32 to 16
x = MaxPooling2D((2, 2))(x)

# Second convolutional block
x = Conv2D(32, (3, 3), activation='relu')(x) # Reduced filters from 64 to 32
x = MaxPooling2D((2, 2))(x)

# Third convolutional block
x = Conv2D(64, (3, 3), activation='relu')(x) # Reduced filters from 128 to 64
x = MaxPooling2D((2, 2))(x)

# Fourth convolutional block
x = Conv2D(64, (3, 3), activation='relu')(x) # Reduced filters from 128 to 64
x = MaxPooling2D((2, 2))(x)

# Flatten and dense layers
x = Flatten()(x)
x = Dense(256, activation='relu')(x) # Reduced units from 512 to 256
outputs = Dense(1, activation='sigmoid')(x)

# Create the model
model_Functional = Model(inputs=inputs, outputs=outputs)
model_Functional.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_Functional.summary()
```

```python
history_seq = model_seq.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    epochs=15, # Increased epochs from 10 to 15
    validation_data=test_generator,
    validation_steps=test_generator.samples // 32
)
```

```python
test_loss, test_acc = model_seq.evaluate(test_generator, verbose=2)
print('\nTest accuracy:', test_acc)
```
