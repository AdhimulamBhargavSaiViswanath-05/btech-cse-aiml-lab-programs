LSTM

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
# Parameters
max_features = 10000
maxlen = 500
```

```python
# Load the dataset
print("Loading data...")
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
print(f"{len(x_train)} train sequences")
print(f"{len(x_test)} test sequences")
```

```python
# Pad sequences to ensure uniform length
print("Padding sequences (samples x time)...")
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

```python
from tensorflow.keras.layers import LSTM

# 4. Build the Sequential LSTM model
model = Sequential([
    # Embedding layer: maps word indices to dense vectors
    Embedding(input_dim=max_features, output_dim=32),

    # LSTM layer: replaced SimpleRNN(32) with LSTM(32)
    # This layer has 'gates' to manage memory over longer sequences
    LSTM(32),

    # Dense output layer: single neuron with sigmoid for binary classification
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

```

```python
# Train the model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)

```

```python
model.summary()
```

```python
# Evaluate the model on the test set
print("\n--- Evaluation on Test Set ---")
results = model.evaluate(x_test, y_test)
print(f"Test Loss: {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f}")
```

RNN

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
# Parameters
max_features = 10000
maxlen = 500
```

```python
# Load the dataset
print("Loading data...")
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
print(f"{len(x_train)} train sequences")
print(f"{len(x_test)} test sequences")
```

```python
# Pad sequences to ensure uniform length
print("Padding sequences (samples x time)...")
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

```python
# Build the Sequential RNN model
model = Sequential([
    # Embedding layer: maps word indices to dense vectors
    Embedding(input_dim=max_features, output_dim=32),
    # SimpleRNN layer: processes sequences of vectors
    SimpleRNN(32),
    # Dense output layer: single neuron with sigmoid for binary classification
    Dense(1, activation='sigmoid')
])
```

```python
# Compile the model
model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# Train the model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=128,
    validation_split=0.2
)

```

```python
# Evaluate the model on the test set
print("\n--- Evaluation on Test Set ---")
results = model.evaluate(x_test, y_test)
print(f"Test Loss: {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f}")
```

```python
import pandas as pd

lstm_test_loss = 0.3897
lstm_test_accuracy = 0.8317

# Dynamically retrieve RNN results from the 'results' variable
rnn_test_loss = results[0] # The 'results' variable from the kernel state contains the last model's evaluation (RNN in this case)
rnn_test_accuracy = results[1]

data = {
    'Model': ['LSTM', 'SimpleRNN'],
    'Test Loss': [lstm_test_loss, rnn_test_loss],
    'Test Accuracy': [lstm_test_accuracy, rnn_test_accuracy]
}

df = pd.DataFrame(data)
print(df.to_markdown(index=False))
```
