### Task01:
- Objective: Create a scalar (rank 0), a vector (rank 1 with 5 elements), and a matrix (rank 2 of size 2×3) using tf.constant(), and print their shape and data type.

```python
import tensorflow as tf

# Scalar (Rank 0)
scalar = tf.constant(5)
print("Scalar:")
print("Value:", scalar)
print("Shape:", scalar.shape)
print("Dtype:", scalar.dtype)
print()

# Vector (Rank 1 with 5 elements)
vector = tf.constant([1, 2, 3, 4, 5])
print("Vector:")
print("Value:", vector)
print("Shape:", vector.shape)
print("Dtype:", vector.dtype)
print()

# Matrix (Rank 2 with shape 2x3)
matrix = tf.constant([[1, 2, 3],
                      [4, 5, 6]])
print("Matrix:")
print("Value:", matrix)
print("Shape:", matrix.shape)
print("Dtype:", matrix.dtype)
```

### Task02:
- Objective: Create a tensor, A, of shape (4, 4) filled with random numbers from a uniform distribution between 10 and 20. Then, create a new tensor, B, filled with zeros but having the same shape and data type as A using tf.zeros_like().

- Key Concept: Initialization functions.

```python
import tensorflow as tf

# Create tensor A with random values between 10 and 20
A = tf.random.uniform(shape=(4, 4), minval=10, maxval=20, dtype=tf.float32)

# Create tensor B with zeros, same shape and dtype as A
B = tf.zeros_like(A)

print("Tensor A:")
print(A)
print("Shape:", A.shape)
print("Dtype:", A.dtype)
print()

print("Tensor B:")
print(B)
print("Shape:", B.shape)
print("Dtype:", B.dtype)
```

# Exercise 3: Reshaping and Indexing Data

Task01:
- Objective: Create a tensor T using tf.range(16). Reshape T into a 4x4 matrix, T_matrix. Then, reshape T_matrix back into a 1D vector, T_vector, allowing TensorFlow to infer the size of the single dimension by using -1.
- Key Concept: tf.reshape() and inferred size (-1).

```python
import tensorflow as tf

# Create a 1D tensor with values from 0 to 15
T = tf.range(16)

# Reshape into a 4x4 matrix
T_matrix = tf.reshape(T, (4, 4))

# Reshape back into a 1D vector using inferred size (-1)
T_vector = tf.reshape(T_matrix, (-1,))

print("Original Tensor T:")
print(T)
print("Shape:", T.shape)
print()

print("Reshaped Tensor T_matrix (4x4):")
print(T_matrix)
print("Shape:", T_matrix.shape)
print()

print("Reshaped Tensor T_vector (1D):")
print(T_vector)
print("Shape:", T_vector.shape)
```

```python
import tensorflow as tf

# Create a 3x4x2 tensor with random integers
D = tf.random.uniform(shape=(3, 4, 2),
                      minval=0,
                      maxval=50,
                      dtype=tf.int32)

print("Tensor D:")
print(D)
print("Shape:", D.shape)
print()

# Access the second slice along the first dimension (axis = 0)
second_slice = D[1]
print("Second slice along axis 0 (D[1]):")
print(second_slice)
print()

# Access a specific element (example: index [1, 2, 0])
element = D[1, 2, 0]
print("Element at index D[1, 2, 0]:")
print(element)
```

# Exercise 4: Matrix Operations and Broadcasting


Task01
- Objective: Define a 3x2 matrix M1 and a 2x4 matrix M2. Calculate the matrix product P = M1 @ M2 using the @ operator or tf.matmul(). Print the shape of the result P (it should be 3x4).
- Key Concept: Matrix Algebra (@).

```python
import tensorflow as tf

# Define matrices
M1 = tf.constant([[1, 2],
                  [3, 4],
                  [5, 6]])   # Shape: (3, 2)

M2 = tf.constant([[1, 2, 3, 4],
                  [5, 6, 7, 8]])  # Shape: (2, 4)

# Matrix multiplication using @ operator
P = M1 @ M2
# Alternatively: P = tf.matmul(M1, M2)

print("Matrix M1:")
print(M1)
print("Shape:", M1.shape)
print()

print("Matrix M2:")
print(M2)
print("Shape:", M2.shape)
print()

print("Product Matrix P (M1 @ M2):")
print(P)
print("Shape:", P.shape)
```

Task02
- Objective: Define a vector V_scale = [2., 2., 2.]. Calculate B_out = M1 * V_scale using element-wise multiplication. Explain why this operation works even though M1 is 3x2 and V_scale is rank 1.
- Key Concept: Implicit Broadcasting.

```python
import tensorflow as tf

# Define matrix M1 (3x2)
M1 = tf.constant([[1., 2.],
                  [3., 4.],
                  [5., 6.]])

# Define scaling vector
V_scale = tf.constant([2., 2., 2.])

# Element-wise multiplication using broadcasting
B_out = M1 * tf.reshape(V_scale, (3, 1))

print("Matrix M1:")
print(M1)
print("Shape:", M1.shape)
print()

print("Scaling Vector V_scale:")
print(V_scale)
print("Shape:", V_scale.shape)
print()

print("Output after Broadcasting (B_out):")
print(B_out)
print("Shape:", B_out.shape)
```

Task03:
- Objective: Define a scalar S = 5.0. Manually use tf.broadcast_to() to expand S into a 3x2 matrix, S_matrix. Verify that M1 + S produces the exact same result as M1 + S_matrix.
- Key Concept: Explicit Broadcasting (tf.broadcast_to).

```python
import tensorflow as tf

# Define matrix M1 (3x2)
M1 = tf.constant([[1., 2.],
                  [3., 4.],
                  [5., 6.]])

# Define scalar
S = tf.constant(5.0)

# Explicitly broadcast scalar to 3x2 matrix
S_matrix = tf.broadcast_to(S, shape=(3, 2))

# Add scalar and broadcasted matrix to M1
result_scalar_add = M1 + S
result_matrix_add = M1 + S_matrix

print("Matrix M1:")
print(M1)
print()

print("Broadcasted Scalar Matrix S_matrix:")
print(S_matrix)
print()

print("Result of M1 + S:")
print(result_scalar_add)
print()

print("Result of M1 + S_matrix:")
print(result_matrix_add)
print()

# Verify both results are identical
print("Are both results equal?",
      tf.reduce_all(result_scalar_add == result_matrix_add).numpy())
```

# Exercise 5: Data Types and Casting

Task01:
- Objective: Define a tensor F_float containing floating-point values: [1.9, 2.1, 3.5, 4.0] (dtype=float32). Use tf.cast() to convert F_float to integer type (tf.int32). Note and explain the conversion outcome (truncation, not rounding).
- Key Concept: Type Casting and Truncation.

```python
import tensorflow as tf

# Define float tensor
F_float = tf.constant([1.9, 2.1, 3.5, 4.0], dtype=tf.float32)

# Cast to int32
F_int = tf.cast(F_float, dtype=tf.int32)

print("Original Float Tensor (F_float):")
print(F_float)
print("Dtype:", F_float.dtype)
print()

print("Casted Integer Tensor (F_int):")
print(F_int)
print("Dtype:", F_int.dtype)
```

Task02:
- Objecctive: Define a tensor B_bool containing booleans: [True, False, True, True]. Cast B_bool to tf.int32. Also, define an integer tensor I_int = [-5, 0, 1] and cast it to tf.bool. Explain the resulting integer/boolean representations.
- Key Concept: Boolean-Integer Mapping.

```python
import tensorflow as tf

# Define boolean tensor
B_bool = tf.constant([True, False, True, True])

# Cast boolean tensor to int32
B_int = tf.cast(B_bool, dtype=tf.int32)

print("Boolean Tensor (B_bool):")
print(B_bool)
print("Dtype:", B_bool.dtype)
print()

print("Casted Integer Tensor (B_int):")
print(B_int)
print("Dtype:", B_int.dtype)
print()

# Define integer tensor
I_int = tf.constant([-5, 0, 1])

# Cast integer tensor to bool
I_bool = tf.cast(I_int, dtype=tf.bool)

print("Integer Tensor (I_int):")
print(I_int)
print("Dtype:", I_int.dtype)
print()

print("Casted Boolean Tensor (I_bool):")
print(I_bool)
print("Dtype:", I_bool.dtype)
```
