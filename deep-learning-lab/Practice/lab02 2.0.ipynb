```python
# load the dataset: /content/drive/MyDrive/DL Lab Practice/datasets/spotify_features.csv
import pandas as pd
spotify_df = pd.read_csv("/content/drive/MyDrive/DL Lab Practice/datasets/spotify_features.csv")
print("The dataset is: \n",spotify_df.head())
```

```python
# Understanding the dataset
print("The information about the spotify dataset is: \n",spotify_df.info())
print("\nThe description about the spotify dataset is: \n",spotify_df.describe())
print("\nThe shape of the spotify dataset is: \n",spotify_df.shape)
print("\nThe columns of the spotify dataset are: \n",spotify_df.columns)
print("\nThe datatypes of the columns of the spotify dataset are: \n",spotify_df.dtypes)
print("\nThe memory usage of the spotify dataset is: \n",spotify_df.memory_usage())
```

```python
# find the missing values
print("The missing values in the spotify dataset are: \n",spotify_df.isnull().sum())
# correct the missing values
spotify_df.fillna(spotify_df.fillna(0), inplace=True)
print("The corrected missing values in the spotify dataset are: \n",spotify_df.isnull().sum())
# input features & target variable
features_spotify = ['danceability', 'energy', 'loudness', 'tempo', 'acousticness', 'speechiness']
target_spotify = 'popularity'
print("The input feature in the spotify dataset are: ", features_spotify)
print("The target variable in the spotify dataset is: ", target_spotify)
```

```python
print("The final dataset is looks like: ", spotify_df.head())
```

```python
# normalize the input features
from sklearn.preprocessing import MinMaxScaler

# input features
scaler_X = MinMaxScaler()

spotify_df[features_spotify] = scaler_X.fit_transform(spotify_df[features_spotify])

print("The normalized input features in the spotify dataset are:\n")
print(spotify_df[features_spotify].head())

# output feature
scaler_y = MinMaxScaler()

spotify_df[[target_spotify]] = scaler_y.fit_transform(spotify_df[[target_spotify]])

print("\nThe normalized target variable in the spotify dataset is:\n")
print(spotify_df[[target_spotify]].head())
```

```python
# EDA
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of target
spotify_df['popularity'].hist(bins=30)
plt.title("23BQ1A4201 - Distribution of Popularity")
plt.show()

print("\n")

# Boxplot to detect outliers
spotify_df[features_spotify].boxplot(figsize=(10,5))
plt.xticks(rotation=45)
plt.title("23BQ1A4201 - Boxplot of Input Features")
plt.show()

print("\n")

# Correlation with target
sns.heatmap(spotify_df[features_spotify + ['popularity']].corr(), annot=True)
plt.title("23BQ1A4201 - Correlation Heatmap")
plt.show()

print("\n")
```

```python
# split the dataset into training and testing
from sklearn.model_selection import train_test_split
X = spotify_df[features_spotify]
y = spotify_df[target_spotify]
print("The shape of the input features is: ", X.shape)
print("The shape of the target variable is: ", y.shape)
# split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
print("The shape of the training set is: ", X_train.shape)
print("The shape of the testing set is: ", X_test.shape)
print("The shape of the target variable in the training set is: ", y_train.shape)
```

```python
# Build the ANN (FFNN) Regression Model Using TensorFlow/Keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout

spotify_model = Sequential()

# Entry Point
spotify_model.add(Input(shape=(6,)))

# Layer 1: Increased neurons for better learning
spotify_model.add(Dense(128, activation='relu'))

# Layer 2: Deeper layer
spotify_model.add(Dense(64, activation='relu'))

# Layer 3: Gradually tapering down
spotify_model.add(Dense(32, activation='relu'))

# Layer 4: Final refinement
spotify_model.add(Dense(16, activation='relu'))

# Output Layer: SIGMOID is best here because your target is scaled 0 to 1
spotify_model.add(Dense(1, activation='sigmoid'))

print("The model summary is: \n")
spotify_model.summary()
```

```python
# Compile the model
from tensorflow.keras.optimizers import Adam

spotify_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']
)

# Show model summary
print("\nThe model summary is (After Compilation): \n")
spotify_model.summary()

# compiled model
print("\nThe model is compiled")
```

```python
# train the model
from tensorflow.keras.callbacks import EarlyStopping


# 4. Training with longer patience
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=20,             # Wait longer to see if it improves
    restore_best_weights=True,
    verbose=1
)

print("Starting Training...")
history_s = spotify_model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=64,           # Increased batch size slightly for stability
    callbacks=[early_stopping],
    verbose=1
)

print("\nTraining Complete!")
print(f"Stopped at epoch: {len(history_s.history['loss'])}")
print(f"Best validation loss: {min(history_s.history['val_loss']):.4f}")
print(f"Best validation MAE: {min(history_s.history['val_mae']):.4f}")
```

```python
# Evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_pred = spotify_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\n===== MODEL PERFORMANCE =====")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R2 Score: {r2:.4f}")
```

```python
import pandas as pd
# Comparison Table
comparison_df = pd.DataFrame({
    "Actual_Popularity": y_test.values.flatten(),
    "Predicted_Popularity": y_pred.flatten()
})

print("\nSample Predictions:")
print(comparison_df.head(10))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(history_s.history['loss'], label='Training Loss')
plt.plot(history_s.history['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.title("Training vs Validation Loss - Spotify ANN")
plt.legend()
plt.grid(True)
plt.show()
```
