```python
# ===== Step 1: Read input files =====
with open('doc1.txt', 'r') as f1, open('doc2.txt', 'r') as f2:
    doc1 = f1.read()
    doc2 = f2.read()


# ===== Step 2: Split text manually =====
def split_text(input_string):
    result = []
    word = ''
    for char in input_string:
        if char in (' ', '\t', '\n'):
            if word:
                result += [word]
                word = ''
        else:
            word += char
    if word:
        result += [word]
    return result


split1 = split_text(doc1)
split2 = split_text(doc2)


# ===== Step 3: Remove punctuation =====
def remove_punctuation(words):
    punctuation_marks = ['.', ',', '!', '?', ';', ':', '"', "'", '(', ')',
                         '[', ']', '{', '}', '-', '_', '/', '\\', '`', '~',
                         '@', '#', '$', '%', '^', '&', '*', '=', '+', '<',
                         '>', '|']
    cleaned_words = []
    for word in words:
        clean_word = ''
        for char in word:
            if char not in punctuation_marks:
                clean_word += char
        if clean_word:
            cleaned_words.append(clean_word)
    return cleaned_words


remove_p1 = remove_punctuation(split1)
remove_p2 = remove_punctuation(split2)


# ===== Step 4: Manual lowercase =====
def manual_lowercase(word):
    char_map = {
        'A': 'a', 'B': 'b', 'C': 'c', 'D': 'd', 'E': 'e', 'F': 'f', 'G': 'g', 'H': 'h', 'I': 'i',
        'J': 'j', 'K': 'k', 'L': 'l', 'M': 'm', 'N': 'n', 'O': 'o', 'P': 'p', 'Q': 'q', 'R': 'r',
        'S': 's', 'T': 't', 'U': 'u', 'V': 'v', 'W': 'w', 'X': 'x', 'Y': 'y', 'Z': 'z'
    }
    result = ''
    for char in word:
        if char in char_map:
            result += char_map[char]
        else:
            result += char
    return result


def lowercase_words(word_list):
    result = []
    for word in word_list:
        result.append(manual_lowercase(word))
    return result


lower1 = lowercase_words(remove_p1)
lower2 = lowercase_words(remove_p2)


# ===== Step 5: Manual length function =====
def get_length(word):
    count = 0
    for _ in word:
        count += 1
    return count


# ===== Step 6: Remove stop words =====
stop_words = {
    "is", "a", "the", "for", "this", "an", "and", "in", "of", "on",
    "to", "its", "by", "has", "how", "we", "it", "with", "due", "since", "over", "from"
}


def remove_stop_words(word_list, stop_words):
    result = []
    for word in word_list:
        if word not in stop_words:
            result.append(word)
    return result


remove_sw1 = remove_stop_words(lower1, stop_words)
remove_sw2 = remove_stop_words(lower2, stop_words)


# ===== Step 7: Manual suffix checker =====
def ends_with(word, suffix):
    if len(word) < len(suffix):
        return False
    for i in range(1, len(suffix) + 1):
        if word[-i] != suffix[-i]:
            return False
    return True


# ===== Step 8: Protected words =====
protected = {
    # From your original
    "universe", "university", "police", "policy",
    "analysis", "analyst", "history", "historic",

    # Nouns ending with 'ing' that aren't verbs
    "spring", "king", "thing", "ceiling", "building", "morning", "evening",

    # Words that look like past tense but aren't meant to stem
    "composed", "supposed", "exposed", "proposed",

    # Words that look like continuous tense but are fixed nouns/adjectives
    "routing", "housing", "painting", "booking",

    # Prevent chopping in 'bringing' irregular case
    "bringing",

    # Other common over-stemming traps
    "news", "species", "series", "means", "physics", "mathematics",
    "economics", "ethics", "politics", "practice", "justice", "notice",
    "service", "advice", "basis", "crisis", "thesis", "oasis",
    "catalyst", "specialist", "specific", "traffic", "arctic",
    "logic", "magic"
}


# ===== Step 9: Irregular verbs dictionary =====
irregular_verbs = {
    "go": {"past": ["went"], "past_participle": ["gone"]},
    "buy": {"past": ["bought"], "past_participle": ["bought"]},
    "bring": {"past": ["brought"], "past_participle": ["brought"]},
    "run": {"past": ["ran"], "past_participle": ["run"]},
    "begin": {"past": ["began"], "past_participle": ["begun"]},
    "drink": {"past": ["drank"], "past_participle": ["drunk"]},
    "eat": {"past": ["ate"], "past_participle": ["eaten"]},
    "give": {"past": ["gave"], "past_participle": ["given"]},
    "take": {"past": ["took"], "past_participle": ["taken"]},
    "write": {"past": ["wrote"], "past_participle": ["written"]},
    "see": {"past": ["saw"], "past_participle": ["seen"]},
    "do": {"past": ["did"], "past_participle": ["done"]},
    "have": {"past": ["had"], "past_participle": ["had"]},
    "make": {"past": ["made"], "past_participle": ["made"]},
    "say": {"past": ["said"], "past_participle": ["said"]},
    "find": {"past": ["found"], "past_participle": ["found"]},
    "think": {"past": ["thought"], "past_participle": ["thought"]},
    "keep": {"past": ["kept"], "past_participle": ["kept"]},
    "feel": {"past": ["felt"], "past_participle": ["felt"]},
    "leave": {"past": ["left"], "past_participle": ["left"]},
    "meet": {"past": ["met"], "past_participle": ["met"]},
    "stand": {"past": ["stood"], "past_participle": ["stood"]},
    "sit": {"past": ["sat"], "past_participle": ["sat"]},
    "speak": {"past": ["spoke"], "past_participle": ["spoken"]},
    "understand": {"past": ["understood"], "past_participle": ["understood"]},
    "win": {"past": ["won"], "past_participle": ["won"]},
    "lose": {"past": ["lost"], "past_participle": ["lost"]},
    "build": {"past": ["built"], "past_participle": ["built"]},
    "pay": {"past": ["paid"], "past_participle": ["paid"]},
    "send": {"past": ["sent"], "past_participle": ["sent"]},
    "sell": {"past": ["sold"], "past_participle": ["sold"]},
    "catch": {"past": ["caught"], "past_participle": ["caught"]},
    "fight": {"past": ["fought"], "past_participle": ["fought"]},
    "teach": {"past": ["taught"], "past_participle": ["taught"]},
    "hold": {"past": ["held"], "past_participle": ["held"]},
    "hear": {"past": ["heard"], "past_participle": ["heard"]},
    "show": {"past": ["showed"], "past_participle": ["shown"]}
}


# ===== Step 10: Tense detection =====
def detect_tense(word):
    w = word.lower()
    for base, forms in irregular_verbs.items():
        if w in forms["past"]:
            return "Past Simple (irregular)"
        if w in forms["past_participle"]:
            return "Past Participle (irregular)"
        if w == base:
            return "Base Form / Present Simple"
    if ends_with(w, "ing") and len(w) > 4:
        return "Present Participle / Continuous"
    if ends_with(w, "ed") and len(w) > 3:
        return "Past Simple or Past Participle (regular)"
    if (ends_with(w, "s") or ends_with(w, "es")) and len(w) > 2:
        return "Present Simple (3rd person singular)"
    return "Base Form / Present Simple"

# apply morphological order: and morphology:-

# ===== Step 11: Manual stemmer =====
def manual_stem(word):
    w = word.lower()
    if w in protected:
        return w
    for base, forms in irregular_verbs.items():
        if w in forms["past"] or w in forms["past_participle"]:
            return base
    length = len(w)
    if ends_with(w, "ies") and length > 3:
        w = w[:-3] + "y"
    elif ends_with(w, "es") and not ends_with(w, "ses") and not ends_with(w, "xes"):
        w = w[:-2]
    elif ends_with(w, "s") and length > 3:
        w = w[:-1]
    if ends_with(w, "ied") and length > 4:
        w = w[:-3] + "y"
    elif ends_with(w, "eed") and length > 3:
        w = w[:-1]
    elif ends_with(w, "ed") and length > 3:
        w = w[:-2]
        if not ends_with(w, "e"):
            w += "e"
    if ends_with(w, "ing") and len(w) > 4:
        w = w[:-3]
        if len(w) > 2 and w[-1] == w[-2]:
            w = w[:-1]
        if not ends_with(w, "e") and w not in protected:
            w += "e"
    if ends_with(w, "tion"):
        w = w[:-4] + "te"
    elif ends_with(w, "sion"):
        w = w[:-4] + "se"
    elif ends_with(w, "ment") or ends_with(w, "ness") or ends_with(w, "less"):
        w = w[:-4]
    elif ends_with(w, "able") or ends_with(w, "ible"):
        w = w[:-4]
    elif ends_with(w, "al") and len(w) > 4:
        w = w[:-2]
    elif ends_with(w, "er") and len(w) > 4:
        w = w[:-2]
    elif ends_with(w, "ly") and len(w) > 3:
        w = w[:-2]
    elif ends_with(w, "ous") and len(w) > 4:
        w = w[:-3]
    elif ends_with(w, "ive") and len(w) > 4:
        w = w[:-3]
    return w


# ===== Step 12: Apply stemming to full document =====
def apply_stemming(word_list):
    result = []
    for word in word_list:
        result.append((manual_stem(word)))
    return result


final_doc1 = apply_stemming(remove_sw1)
final_doc2 = apply_stemming(remove_sw2)





print("\nAfter preprocessing steps, the finalised doc1 is:\n", final_doc1)
print("\nAfter preprocessing steps, the finalised doc2 is:\n", final_doc2)


def build_vocabulary(doc1, doc2):
    vocab = []

    # Step 1: Add unique words from doc1
    for word in doc1:
        is_present = False
        for existing in vocab:
            if existing == word:
                is_present = True
                break
        if not is_present:
            vocab += [word]

    # Step 2: Add unique words from doc2 (if not already in vocab)
    for word in doc2:
        is_present = False
        for existing in vocab:
            if existing == word:
                is_present = True
                break
        if not is_present:
            vocab += [word]

    return vocab


vocabulary = build_vocabulary(final_doc1, final_doc2)
print("\nVocabulary: \n\n", vocabulary)

def binary_vectorize(doc, vocabulary):
    vector = []

    for vocab_word in vocabulary:
        found = False

        # Check manually if vocab_word exists in doc
        for doc_word in doc:
            if doc_word == vocab_word:
                found = True
                break

        if found:
            vector += [1]
        else:
            vector += [0]

    return vectorh
print("\nBinary Vector for Document 2:\n\n", vector2)

def dot_product(v1, v2):
    total = 0
    for i in range(get_length(v1)):
        total += v1[i] * v2[i]
    return total

def magnitude(v):
    total = 0
    for i in range(get_length(v)):
        total += v[i] * v[i]
    # Square root manually using power (0.5)
    return total ** 0.5

def cosine_similarity_manual(v1, v2):
    dot = dot_product(v1, v2)
    mag1 = magnitude(v1)
    mag2 = magnitude(v2)

    if mag1 == 0 or mag2 == 0:
        return 0  # Avoid division by zero

    return dot / (mag1 * mag2)


similarity = cosine_similarity_manual(vector1, vector2)
print("\nCosine Similarity between doc1 and doc2: ", similarity)
```
