# Pre-processing of a Text Document: stop word removal and stemming 

```python
# Step 1: Manual Word Split
def manual_split(text):
    words = []
    word = ""
    for i in range(get_length(text)):
        ch = text[i]
        if ch != ' ' and ch != '\n':
            word += ch
        else:
            if get_length(word) != 0:
                words.append(word)
                word = ""
    if get_length(word) != 0:
        words.append(word)
    return words

# Step 2: Manual Lowercase Conversion
def to_lowercase(word):
    lower_word = ""
    for i in range(get_length(word)):
        ch = word[i]
        if 'A' <= ch <= 'Z':
            lower_word += chr(ord(ch) + 32)
        else:
            lower_word += ch
    return lower_word

# Manual Length Function
def get_length(word):
    count = 0
    for _ in word:
        count += 1
    return count

# Step 3: Manual Stop Word Removal
stop_words = ["is", "a", "the", "for", "in", "of", "are", "that", "on", "with", "to", "and", "as", "an", "at", "by", "this", "from"]

def compare_words(w1, w2):
    if get_length(w1) != get_length(w2):
        return False
    for i in range(get_length(w1)):
        if w1[i] != w2[i]:
            return False
    return True

def remove_stop_words(word_list):
    filtered = []
    for i in range(get_length(word_list)):
        is_stop = False
        for j in range(get_length(stop_words)):
            if compare_words(word_list[i], stop_words[j]):
                is_stop = True
                break
        if not is_stop:
            filtered.append(word_list[i])
    return filtered

# Step 4: Manual Stemming
suffixes = ["ing", "ed", "es", "s", "tion", "ment"]

def ends_with(word, suffix):
    word_len = get_length(word)
    suffix_len = get_length(suffix)
    if suffix_len >= word_len:
        return False
    for i in range(suffix_len):
        if word[word_len - suffix_len + i] != suffix[i]:
            return False
    return True

def remove_suffix(word, suffix_len):
    new_word = ""
    for i in range(get_length(word) - suffix_len):
        new_word += word[i]
    return new_word

def manual_stem(word):
    for i in range(get_length(suffixes)):
        suffix = suffixes[i]
        if ends_with(word, suffix) and get_length(word) > get_length(suffix) + 2:
            return remove_suffix(word, get_length(suffix))
    return word

def stem_words(word_list):
    stemmed = []
    for i in range(get_length(word_list)):
        stemmed.append(manual_stem(word_list[i]))
    return stemmed

# Input sentence
input_sentence = """Computers are computing devices that are used for various computations in different domains."""

# Step 1: Split
print("Original Sentence:")
print(input_sentence)

print("\nStep 1: Manual Split:")
tokens = manual_split(input_sentence)
print(tokens)

# Step 2: Lowercase
print("\nStep 2: Lowercase:")
lowercased = []
for i in range(get_length(tokens)):
    lowercased.append(to_lowercase(tokens[i]))
print(lowercased)

# Step 3: Stop Word Removal
print("\nStep 3: Stop Word Removal:")
filtered = remove_stop_words(lowercased)
print(filtered)

# Step 4: Stemming
print("\nStep 4: Manual Stemming:")
stemmed = stem_words(filtered)
print(stemmed)
```

```python
with open('doc1.txt', 'r') as f1:
    doc1 = f1.read()

def splitting(input_string):
    result = []
    word = ''
    
    for char in input_string:
        if char in (' ', '\t', '\n'):
            if word != '':
                result += [word]
                word = ''
        else:
            word += char

    # Include the last word if there's any
    if word:
        result += [word]
    return result


split1 = splitting(doc1)

# print("Splitting of Doc1: \n",split1)


def manual_lowercase(word):
    lower_word = ''
    for char in word:
        # Check if the character is uppercase (A-Z)
        if 'A' <= char <= 'Z':
            # Convert to lowercase by adding 32 to ASCII value
            lower_char = chr(ord(char) + 32)
        else:
            lower_char = char
        lower_word += lower_char
    return lower_word

def lowercase_words(word_list):
    result = []
    for word in word_list:
        lower = manual_lowercase(word)
        result += [lower]
    return result

lower1 = lowercase_words(split1)

# print("Conversion into Lowercase of Doc1: \n", lower1)

stop_words = ["is", "a", "the", "for", "this", "an", "and", "in", "of", "on",
              "to", "its", "by", "has", "how", "we", "it", "with", "due", "since", "over", "from"]

def remove_stop_words(word_list, stop_words):
    result = []
    for word in word_list:
        is_stop = False
        for stop in stop_words:
            if word == stop:
                is_stop = True
                break
        if not is_stop:
            result += [word] 
    return result

# Example usage
remove_sw1 = remove_stop_words(lower1, stop_words)

# print("After removing stop words from doc1: \n",remove_sw1)
def manual_stem(word):
    suffixes = ['ious', 'ment', 'ing', 'ive', 'ies', 'ly', 'ed', 'es', 's', 's.', '.', 's,']
    for suffix in suffixes:
        suffix_len = len(suffix)
        if len(word) >= suffix_len and word[-suffix_len:] == suffix:
            if len(word) > suffix_len + 1:
                return word[:-suffix_len]
            else:
                return word 
    return word


# Step 2: Apply stemming to a list
def apply_stemming(word_list):
    return [manual_stem(word) for word in word_list]


stemmed1 = apply_stemming(remove_sw1)

# print("After applying the stemming process of doc1: \n",stemmed1)

final_doc1 = stemmed1
print("After preprocessing steps, the finalised doc1 is: \n\n", final_doc1)
```
