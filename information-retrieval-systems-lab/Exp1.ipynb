## Representation of a Text Document in Vector Space Model and Computing Similarity between two documents.

```python
import math
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# --- Configuration and Initial Data ---
STOP_WORDS = set(stopwords.words('english'))  # Load stopwords once

# Main documents
DOC1 = """
ChatGPT, developed by OpenAI, has brought a significant transformation to the field of artificial intelligence since its launch in November 2022. 
Built upon the powerful GPT architecture, it quickly gained popularity due to its ability to generate human-like responses. 
Over time, it has revolutionized how we interact with machines—finding applications in education, customer service, and creative writing.
"""

DOC2 = """
Google Gemini, introduced by Google DeepMind, has made major strides in the AI landscape. 
Launched as a multimodal model, it integrates text, image, and code processing to deliver advanced responses. 
From its initial release to its present state, Gemini has played a crucial role in enhancing productivity and AI-assisted tasks across industries.
"""

DOC3 = """
DeepSeek AI, a rising player in the AI field, has shown promising advancements in understanding complex user inputs. 
Initially developed to provide a lightweight but efficient alternative, it evolved to support powerful code generation and reasoning capabilities. 
Its journey from inception to its current form reflects the broader revolution in open-source and accessible AI tools.
"""

# High similarity short documents
DOCA = "AI is amazing"
DOCB = "AI is super amazing"

# --- Preprocessing Function ---
def preprocess_document(doc_text):
    tokens = word_tokenize(doc_text.lower())
    return [word for word in tokens if word.isalpha() and word not in STOP_WORDS]

# --- TF-IDF & Cosine Similarity Calculation ---
def calculate_tfidf_and_similarity(doc_texts, doc_names):
    preprocessed_docs = [preprocess_document(doc) for doc in doc_texts]
    N = len(preprocessed_docs)

    tf_docs = [Counter(doc) for doc in preprocessed_docs]
    vocabulary = sorted(set(word for tf in tf_docs for word in tf.keys()))

    # DF calculation
    df_scores = {}
    for word in vocabulary:
        df_scores[word] = sum(1 for tf in tf_docs if word in tf)

    # IDF
    idf_values = {
        word: math.log(N / (1 + df_scores[word])) + 1
        for word in vocabulary
    }

    # TF-IDF Vectors
    tfidf_vectors = [
        [tf.get(word, 0) * idf_values[word] for word in vocabulary]
        for tf in tf_docs
    ]

    # Print summary vectors
    def print_vector(doc_name, vector):
        print(f"\nTF-IDF Vector for {doc_name} (First 5 Words):")
        for word, score in zip(vocabulary[:5], vector[:5]):
            print(f"  {word}: {score:.4f}")
        print(f"\nTF-IDF Vector for {doc_name} (Last 5 Words):")
        for word, score in zip(vocabulary[-5:], vector[-5:]):
            print(f"  {word}: {score:.4f}")

    for i in range(len(doc_texts)):
        print_vector(doc_names[i], tfidf_vectors[i])

    # Cosine similarity between every pair
    def cosine_similarity(v1, v2):
        dot = sum(a * b for a, b in zip(v1, v2))
        mag1 = math.sqrt(sum(a * a for a in v1))
        mag2 = math.sqrt(sum(b * b for b in v2))
        return dot / (mag1 * mag2) if mag1 and mag2 else 0.0

    print("\n--- Cosine Similarity Scores ---")
    for i in range(len(doc_texts)):
        for j in range(i + 1, len(doc_texts)):
            sim = cosine_similarity(tfidf_vectors[i], tfidf_vectors[j])
            print(f"{doc_names[i]} vs {doc_names[j]}: {sim:.4f} ({sim * 100:.2f}%)")


# --- Run the main 3-doc experiment ---

print("\n\n===============================================================================")

print("\n\n=== Low Similarity Test Case ===")
main_docs = [DOC1, DOC2, DOC3]
main_names = ["Doc1 - ChatGPT", "Doc2 - Gemini", "Doc3 - DeepSeek"]
calculate_tfidf_and_similarity(main_docs, main_names)

print("\n\n===============================================================================")

# --- Run the high similarity test ---
print("\n\n=== High Similarity Test Case ===")
similar_docs = [DOCA, DOCB]
similar_names = ["DocA - 'AI is amazing'", "DocB - 'AI is super amazing'"]
calculate_tfidf_and_similarity(similar_docs, similar_names)
```

```python
# ------------------------------------
# Step 1: Define Stop Words
# ------------------------------------
stop_words = ["is", "a", "the", "for", "this", "an", "and", "in", "of", "on",
              "to", "its", "by", "has", "how", "we", "it", "with", "due", "since", "over", "from"]

# ------------------------------------
# Step 2: Manual Lowercase Conversion
# ------------------------------------
def to_lowercase(word):
    lowercase_word = ""
    for ch in word:
        if 'A' <= ch <= 'Z':
            lowercase_word += chr(ord(ch) + 32)
        else:
            lowercase_word += ch
    return lowercase_word

# ------------------------------------
# Step 3: Manual Word Splitting
# ------------------------------------
def manual_split(text):
    words = []
    word = ""
    for ch in text:
        if ch != ' ' and ch != '\n':
            word += ch
        else:
            if word != "":
                words.append(word)
                word = ""
    if word != "":
        words.append(word)
    return words

# ------------------------------------
# Step 4: Manual Stop Word Removal
# ------------------------------------

def remove_stop_words_manually(word_list):
    result = []
    for i in range(len(word_list)):
        word_lower = to_lowercase(word_list[i])
        is_stop = False
        for j in range(len(stop_words)):
            if word_lower == stop_words[j]:
                is_stop = True
                break
        if not is_stop:
            result.append(word_lower)  # Keep lowercase version
    return result

# ------------------------------------
# Step 5: Build Vocabulary (unique words from all docs)
# ------------------------------------
def build_vocabulary(docs):
    vocab = []
    for doc in docs:
        for i in range(len(doc)):
            word = doc[i]
            exists = False
            for j in range(len(vocab)):
                if word == vocab[j]:
                    exists = True
                    break
            if not exists:
                vocab.append(word)
    return vocab

# ------------------------------------
# Step 6: Convert Document to Frequency Vector
# ------------------------------------
def document_to_vector(doc, vocab):
    vec = []
    for i in range(len(vocab)):
        count = 0
        for j in range(len(doc)):
            if vocab[i] == doc[j]:
                count += 1
        vec.append(count)
    return vec

# ------------------------------------
# Step 7: Cosine Similarity Manually
# ------------------------------------
def cosine_similarity(vec1, vec2):
    dot_product = 0
    magnitude1 = 0
    magnitude2 = 0
    for i in range(len(vec1)):
        dot_product += vec1[i] * vec2[i]
        magnitude1 += vec1[i] * vec1[i]
        magnitude2 += vec2[i] * vec2[i]
    if magnitude1 == 0 or magnitude2 == 0:
        return 0.0
    sim = dot_product / ((magnitude1**0.5) * (magnitude2**0.5))
    return sim

# ------------------------------------
# Step 8: Main Program
# ------------------------------------

doc1 = """
ChatGPT, developed by OpenAI, has brought a significant transformation to the field of artificial intelligence since its launch in November 2022. 
Built upon the powerful GPT architecture, it quickly gained popularity due to its ability to generate human-like responses. 
Over time, it has revolutionized how we interact with machines—finding applications in education, customer service, and creative writing.
"""

doc2 = """
Google Gemini, introduced by Google DeepMind, has made major strides in the AI landscape. 
Launched as a multimodal model, it integrates text, image, and code processing to deliver advanced responses. 
From its initial release to its present state, Gemini has played a crucial role in enhancing productivity and AI-assisted tasks across industries.
"""

doc3 = """
DeepSeek AI, a rising player in the AI field, has shown promising advancements in understanding complex user inputs. 
Initially developed to provide a lightweight but efficient alternative, it evolved to support powerful code generation and reasoning capabilities. 
Its journey from inception to its current form reflects the broader revolution in open-source and accessible AI tools.
"""

# Preprocess documents
documents = [doc1, doc2, doc3]
processed_docs = []

for doc in documents:
    split_words = manual_split(doc)
    lower_words = [to_lowercase(word) for word in split_words]
    filtered_words = remove_stop_words_manually(lower_words)
    processed_docs.append(filtered_words)

# Build vocabulary
vocab = build_vocabulary(processed_docs)

# Convert each doc to vector
vectors = []
for doc in processed_docs:
    vec = document_to_vector(doc, vocab)
    vectors.append(vec)

# Compute similarities
print("\nCosine Similarities Between Documents:")
print("Doc1 vs Doc2:", cosine_similarity(vectors[0], vectors[1]))
print("Doc1 vs Doc3:", cosine_similarity(vectors[0], vectors[2]))
print("Doc2 vs Doc3:", cosine_similarity(vectors[1], vectors[2]))

```

```python
with open('doc1.txt', 'r') as f1, open('doc2.txt', 'r') as f2:
    doc1 = f1.read()
    doc2 = f2.read()

def splitting(input_string):
    result = []
    word = ''
    
    for char in input_string:
        if char in (' ', '\t', '\n'):
            if word != '':
                result += [word]
                word = ''
        else:
            word += char

    # Include the last word if there's any
    if word:
        result += [word]
    return result


split1 = splitting(doc1)
split2 = splitting(doc2)

# print("Splitting of Doc1: \n",split1)
# print("Splitting of Doc2: \n",split2)


def manual_lowercase(word):
    lower_word = ''
    for char in word:
        # Check if the character is uppercase (A-Z)
        if 'A' <= char <= 'Z':
            # Convert to lowercase by adding 32 to ASCII value
            lower_char = chr(ord(char) + 32)
        else:
            lower_char = char
        lower_word += lower_char
    return lower_word

def lowercase_words(word_list):
    result = []
    for word in word_list:
        lower = manual_lowercase(word)
        result += [lower]
    return result

lower1 = lowercase_words(split1)
lower2 = lowercase_words(split2)

# print("Conversion into Lowercase of Doc1: \n", lower1)
# print("Conversion into Lowercase of Doc2: \n", lower2)


# Manual Length Function
def get_length(word):
    count = 0
    for _ in word:
        count += 1
    return count


stop_words = ["is", "a", "the", "for", "this", "an", "and", "in", "of", "on",
              "to", "its", "by", "has", "how", "we", "it", "with", "due", "since", "over", "from"]

def remove_stop_words(word_list, stop_words):
    result = []
    for word in word_list:
        is_stop = False
        for stop in stop_words:
            if word == stop:
                is_stop = True
                break
        if not is_stop:
            result += [word] 
    return result

# Example usage
remove_sw1 = remove_stop_words(lower1, stop_words)
remove_sw2 = remove_stop_words(lower2, stop_words)

# print("After removing stop words from doc1: \n",remove_sw1)
# print("After removing stop words from doc2: \n",remove_sw2)

def manual_stem(word):
    suffixes = ['ious', 'ment', 'ing', 'ive', 'ies', 'ly', 'ed', 'es', 's', 's.', '.', 's,']
    for suffix in suffixes:
        suffix_len = get_length(suffix)
        if get_length(word) >= suffix_len and word[-suffix_len:] == suffix:
            if get_length(word) > suffix_len + 1:
                return word[:-suffix_len]
            else:
                return word 
    return word

# Step 2: Apply stemming to a list
def apply_stemming(word_list):
    return [manual_stem(word) for word in word_list]

stemmed1 = apply_stemming(remove_sw1)
stemmed2 = apply_stemming(remove_sw2)

# print("After applying the stemming process of doc1: \n",stemmed1)
# print("After applying the stemming process of doc2: \n",stemmed2)

final_doc1 = stemmed1
final_doc2 = stemmed2
print("After preprocessing steps, the finalised doc1 is: \n\n", final_doc1)
print("\nAfter preprocessing steps, the finalised doc2 is: \n\n", final_doc2)

def build_vocabulary(doc1, doc2):
    vocab = []
    
    for word in doc1:
        found = False
        for v in vocab:
            if v == word:
                found = True
                break
        if not found:
            vocab += [word]

    for word in doc2:
        found = False
        for v in vocab:
            if v == word:
                found = True
                break
        if not found:
            vocab += [word]

    return vocab

vocabulary = build_vocabulary(final_doc1, final_doc2)
print("\nVocabulary: \n\n", vocabulary)

def binary_vectorize(doc, vocabulary):
    vector = []
    
    for vocab_word in vocabulary:
        found = False
        
        # Check manually if vocab_word exists in doc
        for doc_word in doc:
            if doc_word == vocab_word:
                found = True
                break
        
        if found:
            vector += [1]
        else:
            vector += [0]
    
    return vector


vector1 = binary_vectorize(final_doc1, vocabulary)
vector2 = binary_vectorize(final_doc2, vocabulary)

print("\nBinary Vector for Document 1:\n\n", vector1)
print("\nBinary Vector for Document 2:\n\n", vector2)

def dot_product(v1, v2):
    total = 0
    for i in range(get_length(v1)):
        total += v1[i] * v2[i]
    return total

def magnitude(v):
    total = 0
    for i in range(get_length(v)):
        total += v[i] * v[i]
    # Square root manually using power (0.5)
    return total ** 0.5

def cosine_similarity_manual(v1, v2):
    dot = dot_product(v1, v2)
    mag1 = magnitude(v1)
    mag2 = magnitude(v2)
    
    if mag1 == 0 or mag2 == 0:
        return 0  # Avoid division by zero
    
    return dot / (mag1 * mag2)


similarity = cosine_similarity_manual(vector1, vector2)
print("\nCosine Similarity between doc1 and doc2: ", similarity)
```
