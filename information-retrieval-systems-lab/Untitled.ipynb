```python
import os
import shutil
import pandas as pd

# Paths (change these as needed)
input_folder = r"D:\Final_Selection_17_Aug_2025\input"        # folder with all photos
csv_file = r"C:/Users/DELL/OneDrive/Documents/file_copy1.csv"    # CSV file containing file_name column
output_folder = r"D:\Final_Selection_17_Aug_2025\output"      # destination folder

# Create output folder if not exists
os.makedirs(output_folder, exist_ok=True)

# Read the CSV (assuming column is named 'file_name')
df = pd.read_csv(csv_file)

# Loop through filenames
for filename in df['file_name']:
    src = os.path.join(input_folder, filename)
    dest = os.path.join(output_folder, filename)

    if os.path.exists(src):
        shutil.copy2(src, dest)   # copy with metadata
        print(f"Copied: {filename}")
    else:
        print(f"Not found: {filename}")

print("✅ Copy process completed.")
```
