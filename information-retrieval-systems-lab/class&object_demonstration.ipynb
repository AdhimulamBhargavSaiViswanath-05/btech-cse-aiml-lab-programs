```python
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

def preprocess_text(text):
    doc = nlp(text)
    tokens = []
    for token in doc:
        # Ignore stopwords and punctuation
        if token.is_stop or token.is_punct:
            continue
        # Lemmatization + lowercasing
        tokens.append(token.lemma_.lower())
    return tokens

# Read text from file
with open("ABS1.txt", "r", encoding="utf-8") as f:
    input_text = f.read()

# Run preprocessing
processed_tokens = preprocess_text(input_text)
print("Preprocessed Tokens:", processed_tokens)
print("\nLength Of Processed Tokens: ", len(processed_tokens))
```

```python
# ---- Comparison ----
set_manual = set(processor.final_processed)
set_spacy = set(processed_tokens)

matches = set_manual.intersection(set_spacy)
only_in_manual = set_manual - set_spacy
only_in_spacy = set_spacy - set_manual

accuracy = len(matches) / len(set_spacy) * 100

print("Total manual tokens:", len(processor.final_processed))
print("Total spaCy tokens:", len(processed_tokens))
print("Matches:", len(matches))
print("Only in manual:", len(only_in_manual))
print("Only in spaCy:", len(only_in_spacy))
print("Accuracy: {:.2f}%".format(accuracy))

print("\nTokens only in manual:\n", only_in_manual)
print("\nTokens only in spaCy:\n", only_in_spacy)
```
