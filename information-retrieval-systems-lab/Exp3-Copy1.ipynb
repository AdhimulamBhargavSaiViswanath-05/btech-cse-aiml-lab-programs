```python
import os
def split_text(input_string):
    result = []
    word = ''

    for char in input_string:
        if char in (' ', '\t', '\n'):
            if word:
                result += [word]
                word = ''
        else:
            word += char

    if word:
        result += [word]

    return result




# print("Splitting of Doc1: \n",split1)
# print("Splitting of Doc2: \n",split2)



def remove_punctuation(words):
    # Define a list of punctuation symbols to remove
    punctuation_marks = ['.', ',', '!', '?', ';', ':', '"', "'", '(', ')', '[', ']', '{', '}', '-', '_', '/', '\\', '`', '~', '@', '#', '$', '%', '^', '&', '*', '=', '+', '<', '>', '|']

    cleaned_words = []
    for word in words:
        clean_word = ''
        for char in word:
            if char not in punctuation_marks:
                clean_word += char
        if clean_word:  # Only add if not empty after cleaning
            cleaned_words.append(clean_word)

    return cleaned_words


# print("After removing punctuation from doc1: \n",remove_p1)
# print("After removing punctuation from doc2: \n",remove_p2)

def manual_lowercase(word):
    result = ''
    # Define a mapping for uppercase to lowercase characters
    char_map = {
        'A': 'a', 'B': 'b', 'C': 'c', 'D': 'd', 'E': 'e', 'F': 'f', 'G': 'g', 'H': 'h', 'I': 'i',
        'J': 'j', 'K': 'k', 'L': 'l', 'M': 'm', 'N': 'n', 'O': 'o', 'P': 'p', 'Q': 'q', 'R': 'r',
        'S': 's', 'T': 't', 'U': 'u', 'V': 'v', 'W': 'w', 'X': 'x', 'Y': 'y', 'Z': 'z'
    }
    for char in word:
        if char in char_map:
            result += char_map[char]
        else:
            result += char
    return result


def lowercase_words(word_list):
    result = []
    for word in word_list:
        lower = manual_lowercase(word)
        result += [lower]
    return result


# print("Conversion into Lowercase of Doc1: \n", lower1)
# print("Conversion into Lowercase of Doc2: \n", lower2)


# Manual Length Function
def get_length(word):
    count = 0
    for _ in word:
        count += 1
    return count


# Define stop words as a set for faster lookup
stop_words = {
    "is", "a", "the", "for", "this", "an", "and", "in", "of", "on",
    "to", "its", "by", "has", "how", "we", "it", "with", "due", "since", "over", "from"
}

# Remove stop words from list
def remove_stop_words(word_list, stop_words):
    result = []
    for word in word_list:
        if word not in stop_words:
            result  += [word]
    return result



# print("After removing stop words from doc1: \n",remove_sw1)
# print("After removing stop words from doc2: \n",remove_sw2)

# Define base word family dictionary
word_families = {
    # Irregular Verbs
    "run": ["ran", "running", "runs"],
    "go": ["went", "gone", "going", "goes"],
    "come": ["came", "coming", "comes"],
    "eat": ["ate", "eating", "eaten", "eats"],
    "take": ["took", "taking", "taken", "takes"],
    "give": ["gave", "giving", "given", "gives"],
    "write": ["wrote", "writing", "written", "writes"],
    "drive": ["drove", "driving", "driven", "drives"],
    "fly": ["flew", "flying", "flown", "flies"],
    "drink": ["drank", "drinking", "drunk", "drinks"],
    "break": ["broke", "breaking", "broken", "breaks"],
    "build": ["built", "building", "builds"],
    "sing": ["sang", "singing", "sung", "sings"],
    "swim": ["swam", "swimming", "swum", "swims"],
    "see": ["saw", "seeing", "seen", "sees"],
    "begin": ["began", "beginning", "begun", "begins"],
    "speak": ["spoke", "speaking", "spoken", "speaks"],
    "buy": ["bought", "buying", "buys"],
    "feel": ["felt", "feeling", "feels"],
    "find": ["found", "finding", "finds"],
    "get": ["got", "getting", "gets"],
    "have": ["had", "having", "has"],
    "hold": ["held", "holding", "holds"],
    "leave": ["left", "leaving", "leaves"],
    "lose": ["lost", "losing", "loses"],
    "pay": ["paid", "paying", "pays"],
    "ride": ["rode", "riding", "ridden", "rides"],
    "sell": ["sold", "selling", "sells"],
    "sit": ["sat", "sitting", "sits"],
    "stand": ["stood", "standing", "stands"],
    "think": ["thought", "thinking", "thinks"],
    "wear": ["wore", "wearing", "worn", "wears"],
    "win": ["won", "winning", "wins"],
    "understand": ["understood", "understanding", "understands"],

    # Regular Verbs
    "play": ["played", "playing", "plays"],
    "talk": ["talked", "talking", "talks"],
    "watch": ["watched", "watching", "watches"],
    "walk": ["walked", "walking", "walks"],
    "work": ["worked", "working", "works"],
    "help": ["helped", "helping", "helps"],
    "call": ["called", "calling", "calls"],
    "cook": ["cooked", "cooking", "cooks"],
    "laugh": ["laughed", "laughing", "laughs"],
    "look": ["looked", "looking", "looks"],
    "live": ["lived", "living", "lives"],
    "open": ["opened", "opening", "opens"],
    "close": ["closed", "closing", "closes"],
    "clean": ["cleaned", "cleaning", "cleans"],
    "move": ["moved", "moving", "moves"],
    "smile": ["smiled", "smiling", "smiles"],
    "love": ["loved", "loving", "loves"],
    "share": ["shared", "sharing", "shares"],
    "travel": ["traveled", "traveling", "travels"],

    # Adjectives / Adverbs to Base
    "beauty": ["beautiful", "beautifully"],
    "happy": ["happier", "happiest", "happily"],
    "sad": ["sadder", "saddest", "sadly"],
    "quick": ["quicker", "quickest", "quickly"],
    "slow": ["slower", "slowest", "slowly"],
    "angry": ["angrier", "angriest", "angrily"],
    "easy": ["easier", "easiest", "easily"],
    "hard": ["harder", "hardest", "hardly"],
    "strong": ["stronger", "strongest", "strongly"],
    "weak": ["weaker", "weakest", "weakly"],

    # Nouns ↔ Plural Forms
    "child": ["children"],
    "person": ["people"],
    "man": ["men"],
    "woman": ["women"],
    "mouse": ["mice"],
    "foot": ["feet"],
    "tooth": ["teeth"],
    "leaf": ["leaves"],
    "life": ["lives"],
    "knife": ["knives"],
    "city": ["cities"],
    "baby": ["babies"],
    "story": ["stories"],
    "book": ["books"],
    "dog": ["dogs"],
    "cat": ["cats"],
    "car": ["cars"],

    # Other Common Derivatives
    "success": ["successful", "successfully", "unsuccessful"],
    "use": ["useful", "useless", "used", "using"],
    "care": ["careful", "carefully", "careless", "carelessly"],
    "hope": ["hopeful", "hopefully", "hopeless"],
    "power": ["powerful", "powerless"],
    "meaning": ["meaningful", "meaningless"],
    "end": ["ending", "ended", "ends"],
    "begin": ["beginning", "began", "begun"],
    "create": ["created", "creating", "creation", "creative"],
    "Combat": ["combating", "combated"]

    #
}


reverse_lookup = {}
for base, forms in word_families.items():
    for form in forms:
        reverse_lookup[form] = base

def manual_stem(word):
    return reverse_lookup.get(word, word)  # Replace if found, else return original

# Step 4: Apply stemming to a list of words
def apply_stemming(word_list):
    return [manual_stem(word) for word in word_list]




# Step 5: Inverted Index with Frequency
def build_inverted_index_with_frequency(documents):
    index = {}  # Dictionary to hold word: {doc_id: frequency, ...}
    for doc_id, doc in enumerate(documents, start=1):
        for word in doc:
            if word not in index:
                index[word] = {}
            if doc_id not in index[word]:
                index[word][doc_id] = 0
            index[word][doc_id] += 1
    return index

# Main Execution

# Path to your folder containing the 50 documents
folder_path = r"C:\Users\DELL\OneDrive\Documents\IRS LAB\documents"

all_documents = []

for i in range(1, 51):  # Loop over all 50 documents
    filename = f"doc{i}.txt"
    filepath = os.path.join(folder_path, filename)
    
    with open(filepath, 'r', encoding='utf-8') as f:
        data = f.read()
    
    # Preprocessing steps
    split_words = split_text(data)
    lowered = lowercase_words(split_words)
    no_stops = remove_stop_words(lowered, stop_words)
    stemmed = apply_stemming(no_stops)
    
    all_documents += [stemmed]

# Build the inverted index with frequencies
inverted_index = build_inverted_index_with_frequency(all_documents)

# Display inverted index
def display_inverted_index(inverted_index):
    print("\n=== Inverted Index ===\n")
    print(f"{'WORD':<20} {'DOCUMENT FREQUENCIES':<100} {'TOTAL':>5}")
    print("-" * 140)

    for word in sorted(inverted_index):
        doc_freqs = inverted_index[word]
        total = sum(doc_freqs.values())

        # Build document frequency list using 'Doc X: Y' format
        doc_freq_list = []
        for doc in doc_freqs:
            entry = "Doc " + str(doc) + ": " + str(doc_freqs[doc])
            doc_freq_list += [entry] 

        # Manually wrap the doc frequencies into lines of max 100 characters
        max_width = 100
        wrapped_lines = []
        current_line = ""

        for entry in doc_freq_list:
            new_line = current_line + (", " if current_line else "") + entry
            if len(new_line) > max_width:
                wrapped_lines += [current_line]
                current_line = entry
            else:
                current_line = new_line
        if current_line:
            wrapped_lines += [current_line]

        # Print first line with the word
        print(f"{word:<20} {wrapped_lines[0]:<100} {total:>5}")

        # Print remaining lines indented
        for line in wrapped_lines[1:]:
            print(f"{'':<20} {line:<100}")

# To count total vocabulary size
vocab_count = 0
for word in inverted_index:
    vocab_count += 1

print("\nTotal Vocabulary Size:", vocab_count)

display_inverted_index(inverted_index)
print("\n")
```
